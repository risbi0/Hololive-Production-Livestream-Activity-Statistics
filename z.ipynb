```python
from dotenv import load_dotenv
import pandas as pd
import json, requests, os

load_dotenv()
HOLODEX_API_KEY = os.getenv('HOLODEX_API_KEY')
```

```python
pd.read_csv('data/data.csv', index_col=[0])
```

```python
df = pd.read_csv('data/details.csv', index_col=[0])
channel_ids = df.loc[df['active'] != 0, 'ch_id'].to_dict()
channel_ids.pop('hololive')

for name, channel_id in channel_ids.items():
    if name != 'ceci': continue
    #print(f'Updating: {name}')
    response = requests.get(f'https://holodex.net/api/v2/channels/UCvN5h1ShZtc7nly3pezRayg/videos?limit=50', 
                            headers={'X-APIKEY': HOLODEX_API_KEY})
    data = json.loads(response.text)
    print(name, channel_id, data)
```
